```python
#CRIANDO SUBCONJUNTO DO DATAFRAME ORIGINAL SELECIONANDO APENAS ONDE NA COLUNA MENSALIDADE O VALOR SEJA O MÁXIMO DA COLUNA
df[df['mensalidade'] == df['mensalidade'].max()][['grau','curso_busca','cidade_busca','uf_busca','mensalidade']]
```

```python
#CURSOS MAIS BARATOS DO BRASIL

df[df['mensalidade'] == df['mensalidade'].min()][['grau','curso_busca','cidade_busca','uf_busca','mensalidade']]
```

```python
#MENSALIDADE MAIS CARA POR UF, DO MAIS CARO PRO MAIS BARATO

df.groupby('uf_busca')['mensalidade'].max().sort_values(ascending=False)
```

```python
#FAZER O MESMO QUE O ANTERIOR MAS MOSTRANDO TODAS COLUNAS (O loc retorna a LINHA de todos os dados e não só os utilizados para filtragem)

df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()].head()
```
