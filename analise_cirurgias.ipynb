```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
caminho = 'cirurgiasprodamb2022.csv'
df = pd.read_csv(caminho, encoding='cp1252', sep=';', usecols=[0,1,3,4,5,6,7])

df.columns = ['ano','mes','complexidade','estabelecimento','procedimento','qtd_procedimento','id']
```

```python
df.head(10)
```

```python
#QUANTIDADE DE CIRURGIAS POR PROCEDIMENTOS
procedimentos = df['procedimento'].value_counts()
procedimentos.head(5)
```

```python
#ANÁLISE AUTOMÁTICA GERAL DOS DADOS
df.describe(include = 'all')
```

```python
#CIRURGIAS POR HOSPITAL
df['estabelecimento'].value_counts()
```

```python
#CRIAR VARIÁVEL APENAS COM INFORMAÇÕES DO HBDF (HOSPITAL DE BASE DO DF)
df_base = df[df['estabelecimento'] == '0010456 - HBDF']
```

```python
#PROCEDIMENTOS NO HOSPITAL DE BASE
proced_base = df_base['procedimento'].value_counts(normalize = True)

proced_base.sample(10)
```

```python
#DIVIDIR DF PELO PROCEDIMENTO EXODONTIA
df_exodontia = df[df['procedimento'].str.contains('EXODONTIA')]
display(df_exodontia)
```

```python
#Verificar qual hospital realizou mais exodontias
df_exodontia['estabelecimento'].value_counts().plot.bar(legend=True)
```
