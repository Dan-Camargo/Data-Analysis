# Análise dados dos filmes exibidos no Brasil de 1995 a 2023

Com base nos dados da ancine (https://www.gov.br/ancine/pt-br/oca/cinema-) fez-se a seguinte análise dos filmes exibidos no brasil.

Arquivo = listagem-de-filmes-brasileiros-lancados-1995-a-2023.csv

```python
#Imports, settings e declarações.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

path = "listagem-de-filmes-brasileiros-lancados-1995-a-2023.xlsx"

#LER E TRANSFORMAR ARQUIVO EM DATAFRAME EXCLUINDO AS LINHAS INFORMATIVAS (NOTAS)
df = pd.read_excel(path, skiprows=1,skipfooter=46)
```

```python
#CHECAR SE DEU CERTO MOSTRANDO AS PRIMEIRAS 5 LINHAS
df.head()
```

```python
df.tail()
```

```python
#CHECAR TIPOS DOS CAMPOS
df.dtypes
```

Aqui vemos que os tipos de dados das colunas não estão formatados de acordo com o que deveriam, como por exemplo a renda que deveria ser reconhecido como FLOAT
... teremos que alterar manualmente para que possamos ter uma análise correta

```python
df['Renda (R$) acumulada'] = pd.to_numeric(df['Renda (R$) acumulada'])
```

De acordo com o erro não poderei alterar a coluna renda por conta da string ND (ERRO) que está no meio dos dados.

```python
#ANALISANDO A COLUNA RENDA PARA ACHAR POSSÍVEIS ERROS

df['Renda (R$) acumulada'].value_counts()
```

Encontrado 25 valores ND na coluna, que não podem ser parseados

```python
#AQUI CRIAREI UM DF CÓPIA ONDE FAREI TODAS ALTERAÇÕES NECESSÁRIAS E JÁ FAREI A ALTERAÇÃO NOS ND DA COLUNA RENDA

df_novo = df.copy()

df_novo['Renda (R$) acumulada'] = df_novo['Renda (R$) acumulada'].replace("ND","0")

df_novo['Renda (R$) acumulada'] = pd.to_numeric(df_novo['Renda (R$) acumulada'])
```

```python
#QUAIS OS FILMES COM MAIOR RENDA ACUMULADA?:
df_novo.nlargest(5, 'Renda (R$) acumulada')
```

```python
#MUDANDO O TIPO DA COLUNA ANO PARA DATE

df_novo['Ano de Lançamento'] = df['Ano de Lançamento'].copy()

#QUAIS OS ANOS COM MAIORES LANÇAMENTOS?

df_novo['Ano de Lançamento'].value_counts().plot.bar()

# Contagem dos valores de 'Ano de Lançamento'
value_counts = df_novo['Ano de Lançamento'].value_counts()

# Definindo cores alternadas: preto e verde escuro
colors = np.array(['#000000', '#006400'])

# Criar o gráfico de barras com largura e espaçamento ajustados
ax = value_counts.plot.bar(color=colors[np.arange(len(value_counts)) % 2], 
                           width=0.8,  # Keep the bar width the same
                           figsize=(12, 6))  # Increase the figure width

# Adicionar rótulos de dados acima das barras
counter = 0
for p in ax.patches:
    # Determine the color based on the counter value
    color = "green" if counter == 1 else "black"
    
    # Annotate the bars
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5),  # Small vertical offset
                textcoords='offset points',
                fontsize=10,  # Font size
                color=color)  # Label color
    
    # Update counter
    counter = 1 - counter

# Exibir o gráfico
plt.show()
```

```python
#SUBSTITUINDO ND POR 0 NA COLUNA PUBLICO ACUMULADO
df_novo['Público acumulado'] = df_novo['Público acumulado'].replace('ND',0)
#AGORA QUE TODAS CÉLULAS SÃO NUMÉRICAS PODEMOS MUDAR O TIPO PARA NUMÉRICO (FLOAT/INT)
df_novo['Público acumulado'] = pd.to_numeric(df_novo["Público acumulado"])
#CHECAR SE FUNCIONOU
df_novo['Público acumulado'].dtypes
```
